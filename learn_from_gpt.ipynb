# 學習 figure

```python
import matplotlib.pyplot as plt

# 創建 Figure
fig = plt.figure(figsize=(8, 6))

# 添加 Axes
ax1 = fig.add_subplot(1, 1, 1)

# 繪圖 - 線圖
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
ax1.plot(x, y)

# 設定標題和標籤
ax1.set_title('Plotting on Axes')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')

plt.show()
```

## 把畫圖功能放入method

```python
import matplotlib.pyplot as plt

def plot_line_diagram(x_array, y_array):
    # 創建 Figure 和 Axes
    fig, ax = plt.subplots()

    # 使用傳入的數組繪製線圖
    ax.plot(x_array, y_array)

    # 添加標題和標籤
    ax.set_title('Line Diagram')
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')

    # 顯示圖表
    plt.show()

# 使用方法
x_array = [0, 1, 2, 3, 4, 5]
y_array = [0, 1, 4, 9, 16, 25]
plot_line_diagram(x_array, y_array)
```

# 連結 matplotlib 和 pandas

在這個範例中，創建了一個名為 plot_line_diagram_from_dataframe 的方法。 \
這個方法接受一個 DataFrame 和兩個column name作為參數，然後使用這些數據來繪製線圖。

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_line_diagram_from_dataframe(df, x_column, y_column):
    # 創建 Figure 和 Axes
    fig, ax = plt.subplots()

    # 使用 DataFrame 的數據繪製線圖
    ax.plot(df[x_column], df[y_column])

    # 添加標題和標籤
    ax.set_title(f'Line Diagram of {y_column} vs {x_column}')
    ax.set_xlabel(x_column)
    ax.set_ylabel(y_column)

    # 顯示圖表
    plt.show()

# 假設你已經有一個 DataFrame
df = pd.DataFrame({
    'x_values': [0, 1, 2, 3, 4, 5],
    'y_values': [0, 1, 4, 9, 16, 25]
})

# 使用方法
plot_line_diagram_from_dataframe(df, 'x_values', 'y_values')
```

# 拓展 pandas & matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_line_diagram_from_dataframe(df, x_column, y_columns, title='Line Diagram', xlabel='', ylabel='', legend_labels=None):
    fig, ax = plt.subplots()

    for y_column in y_columns:
        ax.plot(df[x_column], df[y_column])

    ax.set_title(title)
    ax.set_xlabel(xlabel if xlabel else x_column)
    ax.set_ylabel(ylabel if ylabel else 'Values')

    if legend_labels:
        ax.legend(legend_labels)
    else:
        ax.legend(y_columns)

    plt.show()

# 假設 DataFrame
df = pd.DataFrame({
    'x_values': [0, 1, 2, 3, 4, 5],
    'y1_values': [0, 1, 4, 9, 16, 25],
    'y2_values': [0, 1, 2, 3, 4, 5]
})

# 使用方法
plot_line_diagram_from_dataframe(df, 'x_values', ['y1_values', 'y2_values'], title='Multi-line Plot', legend_labels=['Line 1', 'Line 2'])
```

# 畫出世界🌍地圖的圖形

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 讀取世界地圖數據
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# 選擇一個指標（這裡用人口）來標注顏色
world['color_values'] = world['pop_est'] / 1e6  # 以百萬為單位的人口

# 繪製地圖
fig, ax = plt.subplots(1, figsize=(15, 10))
world.boundary.plot(ax=ax)
world.plot(column='color_values', ax=ax, legend=True,
           legend_kwds={'label': "Population by Country (in millions)"})

plt.title("World Map colored by Population")
plt.show()
```
